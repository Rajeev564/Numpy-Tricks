# How to draw seaborn Barplot/Bar graph

```python
import seaborn as sns
tips_df=sns.load_dataset("tips")
tips_df
```

```python
tips_df.shape
```

# Here we got empty graph because we don't give any parameter to bar graph

```python
sns.barplot(tips_df.total_bill)
plt.show()
            
```

```python
sns.barplot(x=tips_df.day,y=tips_df.total_bill)
plt.show()
```

```python
sns.barplot(x='day', y='total_bill', hue='sex', data=tips_df)
plt.show()
```

```python
order=['Sun','Thur','Fri','Sat']
sns.barplot(x='day', y='total_bill', hue='sex', data=tips_df,order=order)
plt.show()
```

```python
hue_order=['Female','Male']
order=['Sun','Thur','Fri','Sat']
sns.barplot(x='day', y='total_bill', hue='sex', data=tips_df,hue_order=hue_order)
plt.show()
```

# mean,median,max Estimstor parameter

```python
import numpy as np
sns.barplot(x='day', y='total_bill', hue='sex', data=tips_df,estimator=np.mean)
plt.show()
```

```python
import numpy as np
sns.barplot(x='day', y='total_bill', hue='sex', data=tips_df,estimator=np.median)
plt.show()
```

```python
import numpy as np
sns.barplot(x='day', y='total_bill', hue='sex', data=tips_df,estimator=np.max)
plt.show()
```

# Confidence Interval parameter to change length of line of graph

```python
sns.barplot(x='day', y='total_bill', hue='sex', data=tips_df,ci=12)
plt.show()
```

# use n_boot parameter

```python
import numpy as np
sns.barplot(x='day', y='total_bill', hue='sex', data=tips_df,estimator=np.max,n_boot=2)
plt.show()
```

# use of orient parameter(to change horizotal to vertical or vice versa

# here we got error due to x is catagorical value,so to remove this we have to 
# write catagorical value on y axis

```python
import numpy as np
sns.barplot(y='day', x='total_bill', hue='sex', data=tips_df,estimator=np.max,n_boot=2,orient='h')
plt.show()
```

# use of color parameter

```python
import numpy as np
sns.barplot(x='day', y='total_bill', hue='sex', data=tips_df,color='g')
```

# use of palette

```python
import numpy as np
sns.barplot(x='day', y='total_bill', hue='sex', data=tips_df,palette="hot")
```

# saturation (use to make color dark or thin)

```python
import numpy as np
sns.barplot(x='day', y='total_bill', hue='sex', data=tips_df,saturation=0.3)
```
