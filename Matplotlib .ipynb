```python
x1=np.arange(1,11)
x1
y1=2*x
y1
y2=3*x
y2
plt.subplot(1,2,1)
plt.plot(x1,y1,color='g',linewidth=2,linestyle=':')
plt.subplot(1,2,2)
plt.plot(x1,y2,color='r',linestyle=':',linewidth=2)
plt.show()
```

```python
students={
    "Bob":87,
    "Juliya":68,
    "Ranjan":84
    
}
```

```python
names=list(students.keys())
```

```python
value=list(students.values())
```

```python
plt.bar(names,value)
plt.title("the Bar")
plt.xlabel("names")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```

# Horizontal Bar

```python
plt.barh(names,value)
plt.title("the Bar")
plt.xlabel("names")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```
